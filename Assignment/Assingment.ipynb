# Task 1

```python
#Fibonacci Series
n = 22

if n <= 0:
    fibonacci_sequence = []
elif n == 1:
    fibonacci_sequence = [0]
elif n == 2:
    fibonacci_sequence = [0, 1]
else:
    fibonacci_sequence = [0, 1]
    for i in range(2, n):
        next_number = fibonacci_sequence[-1] + fibonacci_sequence[-2]
        fibonacci_sequence.append(next_number)


fibonacci_sequence
```

```python
#Check for Prime Number
num = 18

if num <= 1:
    is_prime = False
elif num <= 3:
    is_prime = True
elif num % 2 == 0 or num % 3 == 0:
    is_prime = False
else:
    is_prime = True
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            is_prime = False
            break
        i += 6

is_prime
```

# Task 3

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Age'].isnull().sum()
```

```python
df.dropna(subset=['Embarked'], inplace=True)
df['Embarked'].isnull().sum()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df.head()
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

```python
av = df.groupby('Pclass')['Fare'].mean()
av
```

```python
common = df['Embarked'].mode()[0]
common
```

```python
survivalRate = df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='mean')
survivalRate
```
